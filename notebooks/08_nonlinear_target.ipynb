```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/student_level_math_data.csv")

# Nonlinear effects
df["math_score"] = (
    40
    + 0.0005 * df["median_income"]
    - 1.5 * df["student_teacher_ratio"]
    - 20 * np.sqrt(df["free_lunch_pct"])
    + 18 * (df["attendance_rate"] ** 2)
    + 12 * np.log1p(df["parent_edu_index"] * 10)
    + np.random.normal(0, 4, len(df))
).clip(0, 100)

# Save updated data
df.to_csv("../data/raw/student_level_math_data.csv", index=False)

df[["math_score"]].describe()
```
