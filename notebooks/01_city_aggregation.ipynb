```python
import pandas as pd

df = pd.read_csv("../data/raw/student_level_math_data.csv")
df.head()
```

```python
city_df = (
    df.groupby(["city", "state"])
      .agg(
          avg_math_score=("math_score", "mean"),
          median_income=("median_income", "median"),
          student_teacher_ratio=("student_teacher_ratio", "mean"),
          free_lunch_pct=("free_lunch_pct", "mean"),
          attendance_rate=("attendance_rate", "mean"),
          parent_edu_index=("parent_edu_index", "mean"),
          student_count=("math_score", "count")
      )
      .reset_index()
)

city_df.head()
```

```python
print("Cities:", city_df.shape[0])
print("Columns:", city_df.shape[1])
print(city_df.describe(include="all"))
```

```python
city_df.to_csv("../data/processed/city_level_math_performance.csv", index=False)
print("Saved: ../data/processed/city_level_math_performance.csv")
```
