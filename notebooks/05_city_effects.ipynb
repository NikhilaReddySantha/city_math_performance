```python
import pandas as pd
import numpy as np

# Load student-level data
df = pd.read_csv("../data/raw/student_level_math_data.csv")

# Create city-level modifiers (structural differences)
city_effects = {
    "New York": 4,
    "Los Angeles": -1,
    "Chicago": -2,
    "Houston": -3,
    "Phoenix": -4,
    "Philadelphia": -1,
    "San Antonio": -3,
    "San Diego": 3,
    "Dallas": -2,
    "San Jose": 5,
    "Austin": 4,
    "Seattle": 5,
    "Denver": 3,
    "Boston": 6,
    "Atlanta": -1
}

# Apply city effects
df["math_score"] = (
    df["math_score"]
    + df["city"].map(city_effects)
).clip(0, 100)

# Re-aggregate to city level
city_df = (
    df.groupby(["city", "state"])
      .agg(
          avg_math_score=("math_score", "mean"),
          median_income=("median_income", "median"),
          student_teacher_ratio=("student_teacher_ratio", "mean"),
          free_lunch_pct=("free_lunch_pct", "mean"),
          attendance_rate=("attendance_rate", "mean"),
          parent_edu_index=("parent_edu_index", "mean"),
          student_count=("math_score", "count")
      )
      .reset_index()
)

# Save updated dataset
city_df.to_csv("../data/processed/city_level_math_performance.csv", index=False)

city_df[["city", "avg_math_score"]].sort_values("avg_math_score", ascending=False)
```
