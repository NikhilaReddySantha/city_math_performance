```python
import numpy as np
import pandas as pd

np.random.seed(42)

N = 120_000

cities = [
    ("New York", "NY"), ("Los Angeles", "CA"), ("Chicago", "IL"),
    ("Houston", "TX"), ("Phoenix", "AZ"), ("Philadelphia", "PA"),
    ("San Antonio", "TX"), ("San Diego", "CA"), ("Dallas", "TX"),
    ("San Jose", "CA"), ("Austin", "TX"), ("Seattle", "WA"),
    ("Denver", "CO"), ("Boston", "MA"), ("Atlanta", "GA")
]

# fixed: sample city/state properly
idx = np.random.randint(0, len(cities), size=N)
city = np.array([cities[i][0] for i in idx])
state = np.array([cities[i][1] for i in idx])

median_income = np.random.normal(65000, 15000, N).clip(20000, 150000)
student_teacher_ratio = np.random.normal(16, 3, N).clip(10, 30)
free_lunch_pct = np.random.beta(2, 5, N)
attendance_rate = np.random.normal(0.94, 0.03, N).clip(0.75, 1.0)
parent_edu_index = np.random.normal(0.6, 0.15, N).clip(0, 1)

noise = np.random.normal(0, 6, N)

math_score = (
    50
    + 0.00035 * median_income
    - 1.2 * student_teacher_ratio
    - 18 * free_lunch_pct
    + 15 * attendance_rate
    + 10 * parent_edu_index
    + noise
).clip(0, 100)

df = pd.DataFrame({
    "city": city,
    "state": state,
    "median_income": median_income,
    "student_teacher_ratio": student_teacher_ratio,
    "free_lunch_pct": free_lunch_pct,
    "attendance_rate": attendance_rate,
    "parent_edu_index": parent_edu_index,
    "math_score": math_score
})

df.head()
```

```python
import os
os.makedirs("../data/raw", exist_ok=True)

df.to_csv("../data/raw/student_level_math_data.csv", index=False)
print("Saved to: ../data/raw/student_level_math_data.csv")
```
