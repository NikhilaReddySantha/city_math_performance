```python
import pandas as pd
from scipy.stats import pearsonr

city_df = pd.read_csv("../data/processed/city_level_math_performance.csv")

def test_corr(x, y, x_name, y_name):
    r, p = pearsonr(x, y)
    print(f"{x_name} vs {y_name}")
    print(f"  Pearson r = {r:.3f}")
    print(f"  p-value   = {p:.4f}")
    print("-" * 40)

# H1: Poverty vs math score
test_corr(city_df["free_lunch_pct"], city_df["avg_math_score"],
          "Free Lunch %", "Avg Math Score")

# H2: Student-teacher ratio vs math score
test_corr(city_df["student_teacher_ratio"], city_df["avg_math_score"],
          "Student-Teacher Ratio", "Avg Math Score")

# H3: Parent education vs math score
test_corr(city_df["parent_edu_index"], city_df["avg_math_score"],
          "Parent Education Index", "Avg Math Score")
```
