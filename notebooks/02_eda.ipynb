```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

city_df = pd.read_csv("../data/processed/city_level_math_performance.csv")

# Correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(city_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
