```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Load data
df = pd.read_csv("../data/processed/city_level_math_performance.csv")

X = df.drop(columns=["city", "state", "avg_math_score"])
y = df["avg_math_score"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

def evaluate(model, name):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    r2 = r2_score(y_test, preds)
    rmse = mean_squared_error(y_test, preds, squared=False)
    print(f"{name:20s} | R2: {r2:.3f} | RMSE: {rmse:.3f}")

# Baselines
evaluate(DummyRegressor(strategy="mean"), "Dummy Mean")
evaluate(LinearRegression(), "Linear Regression")
evaluate(RandomForestRegressor(n_estimators=200, random_state=42),
         "Random Forest")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Load data
df = pd.read_csv("../data/processed/city_level_math_performance.csv")

X = df.drop(columns=["city", "state", "avg_math_score"])
y = df["avg_math_score"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

def evaluate(model, name):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    r2 = r2_score(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))  
    print(f"{name:20s} | R2: {r2:.3f} | RMSE: {rmse:.3f}")

# Baselines
evaluate(DummyRegressor(strategy="mean"), "Dummy Mean")
evaluate(LinearRegression(), "Linear Regression")
evaluate(RandomForestRegressor(n_estimators=200, random_state=42),
         "Random Forest")
```
